```python
no_samples = 500
sample_size = 150
samples = np.random.choice(raw_data['BloodPressure'], size=(no_samples, sample_size), replace=True)
```

```python
sample_mean = np.mean(samples, axis=1)
sample_stds = np.std(samples, axis=1)
sample_percentile = np.percentile(samples, q=95, axis=1)
```

```python
population_mean = raw_data['BloodPressure'].mean()
population_std = raw_data['BloodPressure'].std()
population_percentile = np.percentile(raw_data['BloodPressure'], q=95)
values1=[sample_mean,population_mean]
mean_values1=["sample_mean","population_mean"]
```

```python
plt.hist(sample_mean, alpha=0.5, label='Sample Means',color='pink')
plt.axvline(x=population_mean, color='green', label='Population Mean')
plt.legend()
plt.xlabel("BloodPressure")
plt.ylabel("Frequency")
plt.title('Comparison of Mean BloodPressure')
plt.show()
```

```python
plt.hist(sample_stds, alpha=0.5, label='Sample Stds',color='orange')
plt.axvline(x=population_std, color='green', label='Population Std')
plt.legend()
plt.xlabel("BloodPressure")
plt.ylabel("Frequency")
plt.title('Comparison of Std BloodPressure')
plt.show()
```

```python
plt.hist(sample_percentile, alpha=0.5, label='Sample Percentiles',color='red')
plt.axvline(x=population_percentile, color='green', label='Population Percentile')
plt.legend()
plt.xlabel("BloodPressure")
plt.ylabel("Frequency")
plt.title('Comparison of Percentile BloodPressure')
plt.show()
```
